#### Data Visualization

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.scatterplot(x='active_since_days', y='num_chars', hue='num_chars', data=data)
plt.title('active_since_days vs. num_chars (Seaborn)')
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.plot(data['duration_seconds'], data['num_words'], marker='o', label='Stock market prediction')
plt.xlabel('duration_seconds')
plt.ylabel('num_words')
plt.title('5-Minute Crafts - Youtube ClickBait Titles (Matplotlib)')
plt.legend()
plt.grid()
plt.show()
```

```python
sns.histplot(data["active_since_days"], color="brown")
plt.title("active_since_days")
plt.show()
```

```python
sns.boxplot(y=data["num_words"], color="orange")
plt.title("num_words")
plt.show()
```
