#### Data Visualization

```python
import matplotlib.pyplot as plt

plt.plot(data['StudentID'], data['Name'], marker='o', label=' Student Performance Predictions')
plt.xlabel('StudentID')
plt.ylabel('Name')
plt.title('Student Performance Predictions (Matplotlib)')
plt.legend()
plt.grid()
plt.show()
```

#### Bar Chart

```python
avg_radius_worst = data.groupby('AttendanceRate')['PreviousGrade'].mean()
plt.bar(avg_radius_worst.index, avg_radius_worst)
plt.xlabel('AttendanceRate')
plt.ylabel('Average radius_worst')
plt.title('Average radius_worst by Student Performance Predictions (Matplotlib)')
plt.xticks(rotation=45)
plt.show()
```

#### Scatter Plot

```python
import seaborn as sns
```

```python
sns.scatterplot(x='ParentalSupport', y='FinalGrade', hue='ParentalSupport', data=data)
plt.title('ParentalSupport vs. FinalGrade (Seaborn)')
plt.show()
```

#### HeatMap

```python
correlation_matrix = data[[	'StudentID',  'AttendanceRate', 'StudyHoursPerWeek', 
                           'PreviousGrade', 'ExtracurricularActivities', 'PreviousGrade', 'FinalGrade']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title(' Student Performance Predictions (Seaborn)')
plt.show()
```

#### Plotly Visualization

#### Pair Plot

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv(r"C:\Users\student_performance.csv")

print(data.columns)

sns.pairplot(data, vars=['StudyHoursPerWeek', 'PreviousGrade', 'StudentID', 'AttendanceRate'], hue='ExtracurricularActivities', palette='coolwarm')

plt.suptitle('Pair Plots of Selected Features')  
plt.show()
```

#### Violin Plot

```python
plt.figure(figsize=(8, 6))
sns.violinplot(data=data, x='StudyHoursPerWeek', y='PreviousGrade', palette='coolwarm')
plt.title('Violin Plot: PreviousGrade by StudyHoursPerWeek')
plt.show()
```

#### Bubble Chart

```python
plt.figure(figsize=(8, 6))
plt.scatter(data['StudyHoursPerWeek'], data['PreviousGrade'], s=data['StudentID'] / 10, alpha=0.5, c=data['AttendanceRate'].apply(lambda x: 1 if x == 'M' else 0), cmap='coolwarm')
plt.title('Bubble Chart: Radius vs Texture, Sized by Area')
plt.xlabel('StudyHoursPerWeek')
plt.ylabel('PreviousGrade')
plt.colorbar(label='StudentID (0: Benign, 1: Malignant)')
plt.show()
```

#### Parallel Coordinates Plot

```python
from pandas.plotting import parallel_coordinates
```

```python
subset = data[['StudentID', 'AttendanceRate', 'StudyHoursPerWeek', 'StudentID', 'AttendanceRate']]
plt.figure(figsize=(10, 6))
parallel_coordinates(subset, class_column='AttendanceRate', color=['red', 'blue'])
plt.title('Parallel Coordinates Plot')
plt.show()
```
